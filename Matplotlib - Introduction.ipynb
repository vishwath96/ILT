# How to create a Line Plot

```python
#importing the libraries
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

#defining the dataset
x = np.arange(0,10,0.1)
y = 2*x + 5

#plotting the datapoints
plt.plot(x,y)
plt.show()
```

# Cutomizing the Line Plot

```python


#defining the dataset
x = np.arange(0,10,1)
y = 2*x + 5
#changing the figure
fig=plt.figure(figsize=(10,5))
#plotting the datapoints
plt.plot(x,y,linewidth = 2.0,linestyle = ":", color = 'b',alpha = 0.1,marker ='o')
plt.title("Line Plot Demo")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.legend(['line1'],loc='best')
plt.grid(True)
plt.show()
```

```python
## Creating two line plots in a single graph
#defining the dataset
x = np.arange(0,10,1)
y = 2*x + 5
y1 = 4.5*x + 1
#changing the figure
fig=plt.figure(figsize=(10,5))
#plotting the datapoints
plt.plot(x,y,linewidth = 2.0,linestyle = ":", color = 'b',alpha = 0.5,marker ='o')
plt.plot(x, y1, linewidth = 2.0, linestyle =':', color = 'g', alpha = 0.5, marker = '^')
plt.title("Line Plot Demo")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.legend(['line1', 'line2'],loc='best')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

x = np.arange(0,10,1)
y1 = 2*x+5
y2 = 3*x+10

#Create two or more plots in a fig
plt.subplot(1,2,1)
#PLotting the points
plt.plot(x,y1)
plt.title('Graph1')

plt.subplot(1,2,2)
plt.plot(x,y2)
plt.title('Graph2')
plt.show()
```

# Customizing our Scatter Plot

```python
import matplotlib.pyplot as plt
%matplotlib inline

#Creating the dataset
a =[10,20,30,40,50,60,70,80]
b =[5,3,2,5,6,1,4,2]
x =[1,2,3,4,3,2,1,2]

#Creating the scatter Plot
fig = plt.figure(figsize = (10,5))
plt.scatter(a,b, c='g', s=100,edgecolors='y',marker='o',alpha =0.9)
plt.scatter(a,x,c='r', s=100,edgecolors='b',marker='*',alpha =0.5)
plt.legend(['b','x'],loc='best')
plt.title("Scatter Plot Demo")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.grid(True)
plt.savefig('Scatter Plot.png')
plt.show()
```

# How to create a Box Plot?

```python
import matplotlib.pyplot as plt
%matplotlib inline

#Data Preparation
total = [20,4,1,30,20,10,20,70,30,10]
order = [10,3,1,15,17,2,30,44,2,1]
discount = [30,10,20,5,10,20,50,60,20,45]
data = list([total, order,discount])
#Plotting the data
plt.boxplot(data)
plt.title("Box Plot Demo")
plt.grid(True)
plt.show()

## Anything over 1.5 times of interquartile range
## calculation --  (Q1 - 1.5 * IQR or Q3 + 1.5 * IQR).
```

# How to create a violin plot?

```python
import matplotlib.pyplot as plt
%matplotlib inline

#Data Preparation
total = [20,4,1,30,20,10,20,70,30,10]
order = [10,3,1,15,17,2,30,44,2,1]
discount = [30,10,20,5,10,20,50,60,20,45]
data = list([total, order,discount])
#Plotting the data
plt.violinplot(data, showmeans = True)
plt.title("Violin Plot Demo")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

#Data Preparation
"""
Like box plots, violin plots are used to represent comparison of a variable distribution (or sample distribution) 
across different "categories" (for example, temperature distribution compared between day and night, 
or distribution of car prices compared across different car makers).
"""
total = [20,4,1,30,20,10,20,70,30,10]
order = [10,3,1,15,17,2,30,44,2,1]
discount = [30,10,20,5,10,20,50,60,20,45]
data = list([total, order,discount])
#Plotting the data
plt.violinplot(data, showmeans = True, showmedians = True)
plt.title("Violin Plot Demo")
plt.grid(True)
plt.show()
```

# How to create a Doughnut Chart

```python
import matplotlib.pyplot as plt
%matplotlib inline

#define the dataset
group_names = ["GroupA","GroupB","GroupC"]
group_size = [20,30,50]
size_centre = [5]
a,b,c = [plt.cm.Blues,plt.cm.Reds,plt.cm.Greens]
#Creating the doughnut chart
pie1 = plt.pie(group_size,labels = group_names, radius = 1.5,colors = [a(0.5),b(1.5),c(0.2)])
pie2 = plt.pie(size_centre, radius = 1,colors = 'w')
plt.show()
```

```python
Employee = ['Roshni', 'Shyam', 'Priyanshi',
            'Harshit', 'Anmol']
  
# Setting size in Chart based on 
# given values
Salary = [40000, 50000, 70000, 54000, 44000]
  
# colors
colors = ['blue', 'green', 'yellow', 'orange', 'red']
# explosion
explode = (0.05, 0.05, 0.05, 0.05, 0.05)
  
# Pie Chart
plt.pie(Salary, colors=colors, labels=Employee, autopct='%1.1f%%', pctdistance=0.85, explode=explode)
  
# draw circle
centre_circle = plt.Circle((0, 0), 0.70, fc='white')

##Get current figure
fig = plt.gcf()
  
# Adding Circle in Pie chart
fig.gca().add_artist(centre_circle)
  
# Adding Title of chart
plt.title('Employee Salary Details')
  
# Displaing Chart
plt.show()
```

## Quiver plot

```python
"""
Quiver plot is basically a type of 2D plot which shows vector lines as arrows. 
This type of plots are useful in Electrical engineers to visualize electrical potential 
and show stress gradients in Mechanical engineering.
"""

"""Here x_pos and y_pos are the starting positions of the arrow 
while x_dir and y_dir are the directions of the arrow. 
The below plot contains one quiver arrow starting at x_pos = 0 and y_pos = 0.
The direction of the arrow is pointing towards up and the right at x_dir = 1 and y_dir = 1"""
```

```python
x_pos = 0
y_pos = 0
x_direct = 1
y_direct = 1
 
# Creating plot
fig, ax = plt.subplots(figsize = (10, 5))
ax.quiver(x_pos, y_pos, x_direct, y_direct)
ax.set_title('Quiver plot with one arrow')
 
# Show plot
plt.show()
```

## Stream Plot

```python
"""
Stream plot is basically a type of 2D plot used majorly by physicists to show fluid flow and 2D field gradients
"""
```

```python
import numpy as np
import matplotlib.pyplot as plt
 
# Creating dataset
x = np.arange(0, 10)
y = np.arange(0, 10)
 
# Creating grids
X, Y = np.meshgrid(x, y)
 
# x-component to the right
u = np.ones((10, 10))
 
# y-component zero
v = np.zeros((10, 10))
 
fig = plt.figure(figsize = (12, 7))
 
# Plotting stream plot
plt.streamplot(X, Y, u, v, density = 0.5)
 
# show plot
plt.show()
```
