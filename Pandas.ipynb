```python
import matplotlib.pyplot as plt

y = df['hp']
x = range(32)

plt.plot(x, y)


```

```python
## Area plot

y1 = df['hp']
y2 = df['disp']

x = range(32)

plt.plot(x, y1, linewidth=2.0, color='c') ## Creates a Line Plot
plt.stackplot(x, y1, colors='purple', alpha = 0.7) ## Creates an area plot
plt.plot(x, y2, linewidth = 1.0, color='r')
plt.stackplot(x, y2, colors='black', alpha=0.1)
```

```python
## Bar Plot

y = df['hp']
x = range(32)
x1 = df['Unnamed: 1'].tolist()

fig = plt.figure(figsize=(30,10))
plt.xticks(rotation=90)
plt.bar(x1, y, color='purple', alpha = 0.4)
```

```python
## Horizontal plot
fig = plt.figure(figsize=(20, 15))

plt.barh(x1, y, color='yellow', alpha = 0.7)
```
